```python
df_se = df[df['이름'] == '고세환'].copy()
df_se
```

```python
df_se.info()
```

```python
df_se['Date'] = df_se['등록일시'].dt.date  # Extract the date from the 'Registration Date' column and create a new 'Date' column

df_se_daily_data = df_se.groupby('Date')['전력량1 (Wh)'].sum().reset_index()  # Calculate the daily sum of 'Wattage1 (Wh)' and create a new DataFrame
df_se_sum = df_se.groupby('Date')['조도1 (%)'].sum().reset_index()
df_se_daily_data = df_se_daily_data.merge(df_se_sum, on='Date', how='left')

df_se_daily_data['Ele_Daily Average'] = df_se_daily_data['전력량1 (Wh)'] / 24  # Calculate the daily average by dividing the sum by 24 (number of hours in a day)
df_se_daily_data['lig_Daily Average'] = df_se_daily_data['조도1 (%)'] / 24
df_se_daily_data['이름'] = ['고세환','고세환','고세환','고세환','고세환','고세환','고세환']
df_se_daily_data
```

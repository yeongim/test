```python
df = pd.read_excel('Emotion_Stat_Dataset1.xlsx')
df.tail(30)
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
```

```python
df.info()
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
# 'Year', 'Week', 'Day' 열 추가
df['Year'] = pd.to_datetime(df['Datetime']).dt.isocalendar().year
df['Week'] = pd.to_datetime(df['Datetime']).dt.isocalendar().week
df['Day'] = pd.to_datetime(df['Datetime']).dt.isocalendar().day
```

```python
df.tail(50)
```

```python
df.drop(['Start_Time','End_Time'], axis = 1, inplace = True)
df
```

```python
df.tail(20)
```

```python
week_df = df.groupby(['User', 'Week'])['Negative_Count'].sum().reset_index()
week_df
```

```python
# 'week'에 해당하는 실제 날짜 구하기 -> 한 주의 시작 날짜
# 'Week' 값을 해당 주의 첫 번째 날짜로 변환하여 'Week_Start_Date' 열 추가

week_df['Week_Start_Date'] = pd.to_datetime(df['Year'].astype(str) + week_df['Week'].astype(str) + '1', format='%Y%W%w')
week_df
```
